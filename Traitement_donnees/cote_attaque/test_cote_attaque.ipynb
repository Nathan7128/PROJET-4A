```python
import pandas as pd
import sqlite3
```

```python
connect = sqlite3.connect("../../Databases/raw-database.db")
cursor = connect.cursor()
```

```python
req = cursor.execute(f"SELECT * FROM SKC_cote_attaque_home_team")
res = req.fetchall()
desc = req.description
SKC_cote_attaque_home_team = pd.DataFrame(res)
SKC_cote_attaque_home_team.columns = [i[0] for i in desc]
```

```python
# Importation des données de matching entre les id Skill Corner et Stats Bomb des matches
req = cursor.execute(f"SELECT * FROM matching_matches")
res = req.fetchall()
desc = req.description
matching_matches = pd.DataFrame(res)
matching_matches.columns = [i[0] for i in desc]
```

```python
req = cursor.execute(f"SELECT * FROM SB_matches")
res = req.fetchall()
desc = req.description
SB_matches = pd.DataFrame(res)
SB_matches.columns = [i[0] for i in desc]
```

```python
# Importation des données de matching entre les id Skill Corner et Stats Bomb des équipes
req = cursor.execute(f"SELECT * FROM matching_teams")
res = req.fetchall()
desc = req.description
matching_teams = pd.DataFrame(res)
matching_teams.columns = [i[0] for i in desc]
```

```python
df_home_team = SKC_cote_attaque_home_team.merge(matching_matches).merge(SB_matches[["match_id_SB", "home_team_id_SB"]]).rename({"home_team_id_SB" : "team_id_SB"},
                        axis = 1).merge(matching_teams).drop(["match_id_SB", "team_id_SB"], axis = 1)
```

```python
df_away_team = SKC_cote_attaque_home_team[["match_id_SKC", "seconde_periode", "premiere_periode"]].merge(matching_matches).merge(SB_matches[["match_id_SB", "away_team_id_SB"]]).rename({"away_team_id_SB" : "team_id_SB"},
                        axis = 1).merge(matching_teams).drop(["match_id_SB", "team_id_SB"], axis = 1).rename({"seconde_periode" : "premiere_periode", "premiere_periode" : "seconde_periode"}, axis = 1)
```
