```python
connect = sqlite3.connect("raw-database.db")
cur = connect.cursor()
```

```python
req = cur.execute(f"SELECT * FROM freeze_frames WHERE match_id in (1510819, 1299412, 1160044)")
res = req.fetchall()
desc = req.description
df = pd.DataFrame(res)
df.columns = [i[0] for i in desc]
```

```python
req = cur.execute(f"SELECT * FROM freeze_frames F JOIN events E ON F.event_id = E.id")
res = req.fetchall()
desc = req.description
df = pd.DataFrame(res)
df.columns = [i[0] for i in desc]
```
